# superset , subset, disjoint

```python
a={1,2,3,4,5,6,7,8,9}
b={22,23,4,5,67,8,9}
c={1,2,3,4,5,8,9}
print(a)
print(b)
print(c)
```

```python
a.issuperset(c)
```

```python
a.isdisjoint(b)
```
