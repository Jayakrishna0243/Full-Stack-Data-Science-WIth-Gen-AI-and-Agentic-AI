# Arithmatical opertors

```python
a=20
b=5
print(a+b)
print(a-b)
print(a/b)
print(a//b)
print(a*b)
print(a%b)
print(a**b)
```

# Assignment operators

```python
a = 20
b = 5

a += b
print(a)   

a -= b
print(a)   

a *= b
print(a)   

a /= b
print(a)   

a //= b
print(a)  

a **= b
print(a)    

a %= b
print(a)   
```

# relational operator

```python
a = 20
b = 5

print(a > b)   
print(a < b)   
print(a >= b)  
print(a <= b) 
print(a == b)  
print(a != b)
```
