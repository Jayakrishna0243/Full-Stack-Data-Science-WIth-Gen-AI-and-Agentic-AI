# string slicing

```python
'''string slicing can be divide into two various types are 
1.forward slicing ( start from right to left with 0 indexing)
2. backward slicing(it statrts with left to right with -1 )'''
```

# forword indexing

```python

s='jayakrishna'
s[0:200]
```

```python
s='jayakrishna'
s[2:7]
```

```python
s='jayakrishna'
s[1:200]
```

```python
s='jayakrishna'
s[1:0]
```

# backward indexing

```python
s='jayakrishna'
s[2:-3]
```

```python
s='jayakrishna'
s[0:-5]
```

```python
s='jayakrishna'
s[:200]
```

```python
s='jayakrishna'
s[2:]
```
